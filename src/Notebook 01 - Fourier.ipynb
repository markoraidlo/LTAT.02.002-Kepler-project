```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Parameters
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load in the test data only for now.
from pathlib import Path # Multiplatform compatibility
test_data = pd.read_csv(Path("data") / 'exoTest.csv')
```

```python
# Isolate a time series with an exoplanet and one without
example_exoplanet = test_data[test_data["LABEL"] == 2].drop(columns="LABEL").iloc[2]
example_noplanet  = test_data[test_data["LABEL"] == 1].drop(columns="LABEL").iloc[2]
```

```python
# Plot exoplanet and no planet time series
example_exoplanet.plot()
plt.title("Exoplanet example")
plt.show()

example_noplanet.plot()
plt.title("No exoplanet example")
plt.show()
```

```python
# Compute power spectrum
tf_exoplanet = np.fft.fft(example_exoplanet)
tf_noplanet  = np.fft.fft(example_noplanet)

# Obtain magnitude
tf_exoplanet = np.abs(tf_exoplanet)
tf_noplanet  = np.abs(tf_noplanet)

# Obtain power
tf_exoplanet = pd.Series(tf_exoplanet**2)
tf_noplanet  = pd.Series(tf_noplanet**2)

# Due to symmetry from the analysis, throw away half of it
tf_exoplanet = tf_exoplanet[0:int(len(tf_exoplanet)/2)]
tf_noplanet = tf_noplanet[0:int(len(tf_noplanet)/2)]

# Plot
tf_exoplanet.plot()
plt.title("Exoplanet example")
plt.show()
tf_noplanet.plot()
plt.title("No exoplanet example")
plt.show()
```
